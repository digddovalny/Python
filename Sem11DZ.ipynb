```python
# удаляем ненужные колонки: поле "Name" - не несет смысловой нагрузки для дальнейшей обработки данных
# Cabin - около 80% данных не определены
train_clean = train.drop(columns=['Name', 'Cabin'])  
```

```python
train_clean
```

```python
def inpute_data(df):
    df_copy = df.copy()   # копируем во избежание потери оригинала
    
    # пропущенные значения в колонке Embarked будут заменены самым распространенным по столбцу
    str_vars = list(df_copy.select_dtypes(include=['object']).columns.values)
    for col in str_vars:
        if (df_copy[col].isnull().values.any()):
            df_copy[col].fillna(df_copy[col].mode()[0], inplace=True)
    
    # пропущенные значения в колонке Age будут заменены медианным значением по столбцу
    num_vars = list(df_copy.select_dtypes(include=['int64', 'float64']).columns.values)
    for col in num_vars:
        if (df_copy[col].isnull().values.any()):
            df_copy[col].fillna(df_copy[col].median(), inplace=True)
    
    return df_copy
```

```python
train_full = inpute_data(train_clean)
```

```python
train_full.info()
```

# Замена гендера mail/female на child для age <= 18

```python
def update_data(df):
    df_copy = df.copy()
    df_copy.loc[(df.Age <= 18), ('Sex')] = 'child'
    
    return df_copy
```

```python
train_SexChild = update_data(train_full)
```

```python
train_SexChild.query('Age <= 18')
```

```python
train_sex = train_SexChild.groupby(['Survived', 'Sex'], as_index=False).agg({'Ticket': 'count'}).sort_values('Sex', ascending=False)\
                    .rename(columns={'Ticket': 'count'})
```

```python
train_sex
```

```python
sns.barplot(data=train_sex, x='Sex', y='count', hue='Survived', color='green')
```

Вывод: Кол-во выживших женщин значительно превышает кол-во выживших мужчин, выживаемость детей 50 на 50

```python
sns.catplot('Pclass', data=train_SexChild, hue='Survived', kind='count')
```

Шанс на выживание пассажиров увеличивается с повышением класса

```python
plt.figure(figsize=(20, 10))
sns.catplot(x='Sex', y='Survived', col='Pclass', data=train_SexChild, kind='bar')
```

Шанс на выживание пассажиров зависит от кол-ва членов семьи на борту
